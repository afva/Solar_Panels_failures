```python
import cv2
import matplotlib.pyplot as plt
# Read image
img_path = r"dataset_mod\training\training_con_defecto\DJI_20230801125054_0025_T.JPG"
# show image
fig, ax = plt.subplots()

img = cv2.imread(img_path)

img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
ax.imshow(img, cmap='gray')
plt.show()
```

```python
# show image in different color spaces
plt.figure(figsize=(20,10))
plt.subplot(231)
plt.imshow(img)
plt.title('RGB')
plt.subplot(232)
plt.imshow(cv2.cvtColor(img, cv2.COLOR_BGR2HSV))
plt.title('HSV')
plt.subplot(233)
plt.imshow(cv2.cvtColor(img, cv2.COLOR_BGR2LAB))
plt.title('LAB')
plt.subplot(234)
plt.imshow(cv2.cvtColor(img, cv2.COLOR_BGR2YUV))
plt.title('YUV')
plt.subplot(235)
plt.imshow(cv2.cvtColor(img, cv2.COLOR_BGR2YCrCb))
plt.title('YCrCb')
```
